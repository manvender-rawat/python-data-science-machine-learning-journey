WHY ?

When working on real world data, we deal with missing and incorrect values. So numpy has some special built-in functions using which we can handle missing values efficiently.

1. np.isnan() --> It detects missing values. 

    nan --> Not a number (When calculation fails and data value is missing from dataset).

    It returns a Boolean value.

    Syntax: np.isnan(array_name)

2. np.nan_to_num() --> Used to replace missing value with another value.

    Syntax: np.nan_to_num(array_name, nan=value) [default - 0]

3. np.isinf() --> It is used to detect infinite values.

    Syntax: np.isinf(array_name)

    It returns boolean value.

```python
import numpy as np 

arr = np.array([1, np.nan, 3, np.nan, 5])
print(np.isnan(arr))
```

np.nan == np.nan (cannot be compared directly)

```python
cleaned_arr = np.nan_to_num(arr)
print(cleaned_arr)
```

```python
arr1 = np.array([1, 2, np.inf, 3, -np.inf, 4])
print(np.isinf(arr1))
```

To replace infinite values from finite values 

```python
cleaned_arr1 = np.nan_to_num(arr1, posinf=1000, neginf=-1000)
print(cleaned_arr1)
```
